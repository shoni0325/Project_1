### 투자액 -> 비율

```python
import re

# 변환할 열 리스트
columns_to_convert = ['투자액(국내설비)', '투자액(R&D)', '투자액(해외)', '투자액(기타)', '투자액(합계)']

# 각 열을 숫자형으로 변환
for column in columns_to_convert:
    # ,와 . 제거 후 숫자로 변환
    df[column] = df[column].apply(lambda x: re.sub(r'[,\.]', '', x)).astype(float)
```

```python
df['투자액(국내설비)'] = (df['투자액(국내설비)'] / df['투자액(합계)'])*100
df['투자액(R&D)'] = (df['투자액(R&D)'] / df['투자액(합계)'])*100
df['투자액(해외)'] = (df['투자액(해외)'] / df['투자액(합계)'])*100
df['투자액(기타)'] = (df['투자액(기타)'] / df['투자액(합계)'])*100

df.rename(columns={'투자액(국내설비)': '투자액비(국내설비)'}, inplace=True)
df.rename(columns={'투자액(R&D)': '투자액비(RnD)'}, inplace=True)
df.rename(columns={'투자액(해외)': '투자액비(해외)'}, inplace=True)
df.rename(columns={'투자액(기타)': '투자액비(기타)'}, inplace=True)
```
