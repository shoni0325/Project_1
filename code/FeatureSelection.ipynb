# Wrapper Method

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

> unsampled data

```python
# 독립변수, 종속변수 설정
X = us_df.iloc[:,:-1]
y = us_df['벤처캐피털(투자조합포함) 투자 경험여부']

# 변수 선택기 (SequentialFeatureSelector)
selector = SequentialFeatureSelector(estimator=LogisticRegression(), n_features_to_select=15, direction='backward', scoring='f1', cv=5, n_jobs=-1)

# 변수 선택 수행
selector.fit(X, y)

# 선택된 변수의 인덱스
selected_features = selector.get_support(indices=True)

# 선택된 변수 출력
for i in selected_features:
    print(X.columns[i])
```

```python
selected_columns = X.columns[selected_features]
selected_variables = list(selected_columns)
```

> sampled data

```python
# 독립변수, 종속변수 설정
X = mm_df.iloc[:,:-1]
y = mm_df['벤처캐피털(투자조합포함) 투자 경험여부']

# 변수 선택기 (SequentialFeatureSelector)
selector = SequentialFeatureSelector(estimator=LogisticRegression(), n_features_to_select=15, direction='backward', scoring='f1', cv=5, n_jobs=-1)

# 변수 선택 수행
selector.fit(X, y)

# 선택된 변수의 인덱스
selected_features_1 = selector.get_support(indices=True)

# 선택된 변수 출력
for i in selected_features_1:
    print(X.columns[i])
```

```python
selected_columns_1 = X.columns[selected_features_1]
selected_variables_1 = list(selected_columns_1)
```

```python
wrapper_result = pd.DataFrame()

wrapper_result['unsampled']  = selected_variables
wrapper_result['mmsampled'] = selected_variables_1

wrapper_result
```

# Embedded Method

### Lasso

> unsampled data

```python
from sklearn.linear_model import Lasso

X = us_df.iloc[:,:-1]
y = us_df['벤처캐피털(투자조합포함) 투자 경험여부']

# Lasso 모델 생성 및 훈련
lasso = Lasso(alpha=0.5)
lasso.fit(X, y)

# 계수 재조정 및 DF 생성
coefficients = lasso.coef_.reshape(1, -1)

coef_df = pd.DataFrame(data=coefficients, columns=X.columns)
coef_df
```

```python
# 0이 아닌 계수를 가진 변수의 컬럼명 추출
selected_columns = list(X.columns[lasso.coef_ != 0])
len(selected_columns)
```

> sampled data

```python
X = mm_df.iloc[:,:-1]
y = mm_df['벤처캐피털(투자조합포함) 투자 경험여부']

# Lasso 모델 생성 및 훈련
lasso = Lasso(alpha=0.012)
lasso.fit(X, y)

# 계수 재조정 및 DF 생성
coefficients_1 = lasso.coef_.reshape(1, -1)

coef_df_1 = pd.DataFrame(data=coefficients_1, columns=X.columns)
coef_df_1
```

```python
# 0이 아닌 계수를 가진 변수의 컬럼명 추출
selected_columns_1 = list(X.columns[lasso.coef_ != 0])
len(selected_columns_1)
```

```python
lasso_result = pd.DataFrame()

lasso_result['unsampled'] = selected_columns
lasso_result['mmsampled'] = selected_columns_1
lasso_result
```

### SelectFromModel

> unsampled data

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression, ElasticNet
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 데이터 설정
X = us_df.iloc[:,:-1]
y = us_df['벤처캐피털(투자조합포함) 투자 경험여부']

# 특성 선택을 저장할 DF 생성
us_df_select = pd.DataFrame()
us_df_select.index = X.columns

# ---------------------------------------------------------------------

# DecisionTree 분류기 모델 생성 및 훈련
selector = SelectFromModel(estimator=DecisionTreeClassifier(), threshold=0.1).fit(X, y)
us_df_select['DT_0.1'] = selector.get_support()

selector = SelectFromModel(estimator=DecisionTreeClassifier(), threshold=0.2).fit(X, y)
us_df_select['DT_0.2'] = selector.get_support()

selector = SelectFromModel(estimator=DecisionTreeClassifier(), threshold=0.3).fit(X, y)
us_df_select['DT_0.3'] = selector.get_support()

# RandomForest 분류기 모델 생성 및 훈련
selector = SelectFromModel(estimator=RandomForestClassifier(), threshold=0.1).fit(X, y)
us_df_select['RF_0.1'] = selector.get_support()

selector = SelectFromModel(estimator=RandomForestClassifier(), threshold=0.2).fit(X, y)
us_df_select['RF_0.2'] = selector.get_support()

selector = SelectFromModel(estimator=RandomForestClassifier(), threshold=0.3).fit(X, y)
us_df_select['RF_0.3'] = selector.get_support()

# L1 규제를 적용한 특성 선택 수행
selector = SelectFromModel(estimator=LogisticRegression(penalty='l1', solver='liblinear', C=0.01)).fit(X, y)
us_df_select['L1_0.01'] = selector.get_support()

selector = SelectFromModel(estimator=LogisticRegression(penalty='l1', solver='liblinear', C=0.05)).fit(X, y)
us_df_select['L1_0.05'] = selector.get_support()

selector = SelectFromModel(estimator=LogisticRegression(penalty='l1', solver='liblinear', C=0.1)).fit(X, y)
us_df_select['L1_0.1'] = selector.get_support()

# L2 규제를 적용한 특성 선택 수행
selector = SelectFromModel(estimator=LogisticRegression(penalty='l2', solver='liblinear', C=0.01)).fit(X, y)
us_df_select['L2_0.01'] = selector.get_support()

selector = SelectFromModel(estimator=LogisticRegression(penalty='l2', solver='liblinear', C=0.05)).fit(X, y)
us_df_select['L2_0.05'] = selector.get_support()

selector = SelectFromModel(estimator=LogisticRegression(penalty='l2', solver='liblinear', C=0.1)).fit(X, y)
us_df_select['L2_0.1'] = selector.get_support()

# ElasticNet 규제를 적용한 특성 선택 수행
selector = SelectFromModel(estimator=ElasticNet(alpha=0.01, l1_ratio=0.5)).fit(X, y)
us_df_select['elasticnet_0.01'] = selector.get_support()

selector = SelectFromModel(estimator=ElasticNet(alpha=0.05, l1_ratio=0.5)).fit(X, y)
us_df_select['elasticnet_0.05'] = selector.get_support()

selector = SelectFromModel(estimator=ElasticNet(alpha=0.1, l1_ratio=0.5)).fit(X, y)
us_df_select['elasticnet_0.1'] = selector.get_support()


us_df_select['true_sum'] = us_df_select.sum(axis=1)

us_df_select
```

```python
top_indices = us_df_select['true_sum'].nlargest(15).index.tolist()
top_indices
```
