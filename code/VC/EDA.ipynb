### Chi2 & T-test

> Chi 2

* 카이제곱 검정 조건
    - 종속변인은 범주형 자료여야 한다.
    - 기대빈도가 5이하인 셀이 전체의 20%가 넘지 않아야 한다
    - 각 칸의 빈도는 다른 칸의 빈도와 독립적이어야 한다

```python
# 범주형 변수와 타겟변수로 새 데이터프레임 생성
cat_df = df.iloc[:, :21]
cat_df['Exit'] = df['Exit']
```

```python
cat_df.columns
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 독립변수, 종속변수 설정
X = cat_df.iloc[:, :-1]
y = cat_df['Exit']

# 독립변수와 종속변수 간의 카이제곱 검정 수행
chi2_scores = []

p_values = []
for column in X.columns:
    # 교차 테이블 생성
    contingency_table = pd.crosstab(X[column], y)
    chi2, p_value, dof, expected_freq = chi2_contingency(contingency_table)
    p_values.append(p_value)
    chi2_scores.append((column, chi2))

# p-value가 0.05보다 작은 값을 출력하는 데이터프레임 생성
result_df = pd.DataFrame({'Variable': X.columns, 'p-value': p_values})
filtered_df_chi = result_df[result_df['p-value'] < 0.05]

filtered_df_chi
```

#### 정규성 검정

```python
from scipy.stats import shapiro, anderson, kstest,jarque_bera, normaltest,anderson_ksamp
```

```python
# 수치, 비율형 변수와 타겟변수로 새 데이터프레임 생성
cont_df = df.iloc[:,21:-1]
cont_df['Exit'] = df['Exit']
```

```python
# 데이터 분할
group_0 = cont_df[cont_df['Exit'] == 0]
group_1 = cont_df[cont_df['Exit'] == 1]
```

```python
def normality_test(df, tool='Shapiro'):

    result_dict = {'col':[], 'p_value':[], 't_stat':[]}

    if tool == 'Shapiro':
        for i in df.columns:
            statistic, p_value = shapiro(df[i])
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)
    
    elif tool == 'k_s':
        for i in df.columns:
            statistic, p_value = kstest(df[i], 'norm')
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)

    elif tool == 'jarque_bera':
        for i in df.columns:
            statistic, p_value = jarque_bera(df[i])
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)


    elif tool == 'anderson':

        result_dict = {'col' : [], 'critical_values' : [], 't_stat' : []}

        for i in df.columns:
            result = anderson(df[i])
            result_dict['col'].append(i)
            result_dict['critical_values'].append(result.critical_values[2])
            result_dict['t_stat'].append(result.statistic)

            test = pd.DataFrame(result_dict)

    return test
```

### Student t-test

```python
# 등분산 충족하는 변수

# filtered_df['Variable']에 포함된 종속변수 추출
selected_variables = filtered_df['Variable'].tolist()
selected_variables.append('Exit')  # 종속변수 추가

# 선택된 변수만으로 구성된 새 데이터프레임 생성
st_ttest_df = cont_df[selected_variables].copy()

# 새 데이터프레임 출력
st_ttest_df
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# 종속변수와 그룹 0의 데이터 추출
group_0 = st_ttest_df[st_ttest_df['Exit'] == 0]
group_0 = group_0.drop(columns='Exit')

# 종속변수와 그룹 1의 데이터 추출
group_1 = st_ttest_df[st_ttest_df['Exit'] == 1]
group_1 = group_1.drop(columns='Exit')

# 변수별로 Student t-test 수행
p_values = {}
for column in st_ttest_df.columns[:-1]:
    t_value, p_value = ttest_ind(group_0[column], group_1[column])
    p_values[column] = p_value

# 결과를 데이터프레임으로 변환
result_df_st = pd.DataFrame.from_dict(p_values, orient='index', columns=['p-value'])

result_df_st.sort_values('p-value', ascending=True, inplace=True)
result_df_st.reset_index(inplace=True)
result_df_st.rename(columns={'index' : 'Variable'}, inplace=True)
result_df_st
```

```python
# 등분산 충족하지 않은 변수

excluded_variables = ['자기자본비율',
 '학력별 인력구성비(석사)',
 '창업 당시 창업자 실무경험 년수(년)',
 '종사자생산성',
 'ROE',
 '벤처확인제도 혜택(평균)',
 '경영상 애로사항(개수)',
 '투자액비(국내설비)',
 '매출구조비율(B2C)'
 ]

# 선택된 변수만으로 구성된 새 데이터프레임 생성
new_df = cont_df.drop(excluded_variables, axis=1)

new_df['Exit'] = cont_df['Exit']
# new_df.drop('학력별 인력구성비(대졸)', axis=1, inplace=True)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# 독립변수, 종속변수 설정
X = new_df.drop('Exit', axis=1)
y = new_df['Exit']

# Welch's t-test를 통한 변수 선택
selected_features = []

for feature in X.columns:
    group_0 = X[y == 0][feature]
    group_1 = X[y == 1][feature]

    t_statistic, p_value = ttest_ind(group_0, group_1, equal_var=False)

    if p_value < 0.05:  # 유의수준 0.05로 설정
        selected_features.append({'Variable': feature, 'p-value': p_value})

# 결과를 데이터프레임으로 만들어 출력
result_df_wt = pd.DataFrame(selected_features)

result_df_wt.reset_index()
result_df_wt
```
