# 수치형 변수들

```python
cat_df = df.iloc[:,0:9]
cat_df['Exit'] = df['Exit']
```

```python
cat_df_a = cat_df[cat_df['Exit']==0]
cat_df_b = cat_df[cat_df['Exit']==1]
```

```python
cat_a_desc = cat_df_a.describe()

cat_a_desc = pd.DataFrame(cat_a_desc)
cat_a_desc
```

```python
cat_b_desc = cat_df_b.describe()

cat_b_desc = pd.DataFrame(cat_b_desc)
cat_b_desc
```

> 시장점유율(퍼센트)

```python
market_share = pd.DataFrame(index=cat_b_desc.index)

market_share['주력제품 해외 시장점유율(퍼센트)_0'] = cat_a_desc['주력제품 해외 시장점유율(퍼센트)']
market_share['주력제품 해외 시장점유율(퍼센트)_1'] = cat_b_desc['주력제품 해외 시장점유율(퍼센트)']
market_share['주력제품 국내 시장점유율(퍼센트)_0'] = cat_a_desc['주력제품 국내 시장점유율(퍼센트)']
market_share['주력제품 국내 시장점유율(퍼센트)_1'] = cat_b_desc['주력제품 국내 시장점유율(퍼센트)']

market_share
```

> 매출구조비율

```python
sales_struc = pd.DataFrame(index=cat_b_desc.index)

sales_struc['매출구조비율(중소기업)_0'] = cat_a_desc['매출구조비율(중소기업)']
sales_struc['매출구조비율(중소기업)_1'] = cat_b_desc['매출구조비율(중소기업)']
sales_struc['매출구조비율(글로벌)_0'] = cat_a_desc['매출구조비율(글로벌)']
sales_struc['매출구조비율(글로벌)_1'] = cat_b_desc['매출구조비율(글로벌)']

sales_struc
```

```python
financial_rate = pd.DataFrame(index=cat_b_desc.index)

financial_rate['총자산회전율_0'] = cat_a_desc['총자산회전율']
financial_rate['총자산회전율_1'] = cat_b_desc['총자산회전율']
financial_rate['자기자본비율_0'] = cat_a_desc['자기자본비율']
financial_rate['자기자본비율_1'] = cat_b_desc['자기자본비율']
financial_rate['부채비율_0'] = cat_a_desc['부채비율']
financial_rate['부채비율_1'] = cat_b_desc['부채비율']
financial_rate['순이익률_0'] = cat_a_desc['순이익률']
financial_rate['순이익률_1'] = cat_b_desc['순이익률']

financial_rate
```
