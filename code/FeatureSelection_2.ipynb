## Chi 2 & T-test

#### Chi 2

```python
# 범주형 변수와 타겟변수로 새 데이터프레임 생성
cat_df = df.iloc[:, :22]
cat_df['벤처캐피털(투자조합포함) 투자 경험여부'] = df['벤처캐피털(투자조합포함) 투자 경험여부']
```

```python
from scipy.stats import chi2_contingency

# 독립변수, 종속변수 설정
X = cat_df.iloc[:,:-1]
y = cat_df['벤처캐피털(투자조합포함) 투자 경험여부']

# 독립변수와 종속변수 간의 카이제곱 검정 수행
chi2_scores = []

for column in X.columns:
    # 교차 테이블 생성
    contingency_table = pd.crosstab(X[column], y)
    chi2, p_value, dof, expected_freq = chi2_contingency(contingency_table)
    # 카이제곱 통계량(chi2), p-value(p_value), 자유도(dof), 예상빈도(expected_freq)
    print(p_value)
    chi2_scores.append((column, chi2))

# 카이제곱 검정 결과를 기준으로 변수 정렬
sorted_features = sorted(chi2_scores, key=lambda x : x[1], reverse=True)

# 상위 10개의 변수 선택
# 카이제곱 통계량이 클 수록 독립변수와 종속 변수 간의 관련성이 강하다고 본다
sorted_features = sorted_features[:9]

# 선택된 변수 출력
print('Selected Features : ', selected_features)
```
