## Chi 2 & T-test

#### Chi 2

* 카이제곱 검정 조건
    * 종속변인은 범주형 자료여야 한다.
    * 기대빈도가 5이하인 셀이 전체의 20%가 넘지 않아야 한다.
    * 각 칸의 빈도는 다른 칸의 빈도와 독립적이어야 한다.

```python
# 범주형 변수와 타겟변수로 새 데이터프레임 생성
cat_df = df.iloc[:, :22]
cat_df['벤처캐피털(투자조합포함) 투자 경험여부'] = df['벤처캐피털(투자조합포함) 투자 경험여부']
```

```python
cat_df.columns
```

```python
from scipy.stats import chi2_contingency

# 기대빈도가 5 이하인 항목의 비율을 저장할 딕셔너리
expected_freq_5_ratio = {}

# 각 독립 변수에 대해 기대빈도 계산 및 비율 확인
for column in cat_df.columns[:-1]:
    # 교차 테이블 생성
    contingency_table = pd.crosstab(cat_df[column], cat_df['벤처캐피털(투자조합포함) 투자 경험여부'])
    
    # 카이제곱 검정 수행
    chi2, p_value, dof, expected_freq = chi2_contingency(contingency_table)
    
    # 기대빈도가 5 이하인 항목의 비율 계산
    expected_freq_5 = (expected_freq <= 5).mean()
    
    # 결과 저장
    expected_freq_5_ratio[column] = expected_freq_5

# 결과 출력
for column, ratio in expected_freq_5_ratio.items():
    print(f"변수 '{column}'의 기대빈도가 5 이하인 항목 비율: {ratio}")
```

> 정규성 검정

```python
from scipy.stats import shapiro, anderson, kstest,jarque_bera, normaltest,anderson_ksamp
```

```python
def normality_test(df, tool='Shapiro'):

    result_dict = {'col':[], 'p_value':[], 't_stat':[]}

    if tool == 'Shapiro':
        for i in df.columns:
            statistic, p_value = shapiro(df[i])
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)
    
    elif tool == 'k_s':
        for i in df.columns:
            statistic, p_value = kstest(df[i], 'norm')
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)

    elif tool == 'jarque_bera':
        for i in df.columns:
            statistic, p_value = jarque_bera(df[i])
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)


    elif tool == 'anderson':

        result_dict = {'col' : [], 'critical_values' : [], 't_stat' : []}

        for i in df.columns:
            result = anderson(df[i])
            result_dict['col'].append(i)
            result_dict['critical_values'].append(result.critical_values[2])
            result_dict['t_stat'].append(result.statistic)

            test = pd.DataFrame(result_dict)

    return test
```

```python
# 등분산 충족하는 변수

# filtered_df['Variable']에 포함된 종속변수 추출
selected_variables = filtered_df['Variable'].tolist()
selected_variables.append('벤처캐피털(투자조합포함) 투자 경험여부')  # 종속변수 추가

# 선택된 변수만으로 구성된 새 데이터프레임 생성
st_ttest_df = cont_df[selected_variables].copy()

# 새 데이터프레임 출력
st_ttest_df
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# 종속변수와 그룹 0의 데이터 추출
group_0 = st_ttest_df[st_ttest_df['벤처캐피털(투자조합포함) 투자 경험여부'] == 0]
group_0 = group_0.drop(columns='벤처캐피털(투자조합포함) 투자 경험여부')

# 종속변수와 그룹 1의 데이터 추출
group_1 = st_ttest_df[st_ttest_df['벤처캐피털(투자조합포함) 투자 경험여부'] == 1]
group_1 = group_1.drop(columns='벤처캐피털(투자조합포함) 투자 경험여부')

# 변수별로 Student t-test 수행
p_values = {}
for column in st_ttest_df.columns[:-1]:
    t_value, p_value = ttest_ind(group_0[column], group_1[column])
    p_values[column] = p_value

# 결과를 데이터프레임으로 변환
result_df_st = pd.DataFrame.from_dict(p_values, orient='index', columns=['p-value'])

result_df_st.sort_values('p-value', ascending=True, inplace=True)
result_df_st.reset_index(inplace=True)
result_df_st.rename(columns={'index' : 'Variable'}, inplace=True)
result_df_st
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# 독립변수, 종속변수 설정
X = new_df.drop('벤처캐피털(투자조합포함) 투자 경험여부', axis=1)
y = new_df['벤처캐피털(투자조합포함) 투자 경험여부']

# Welch's t-test를 통한 변수 선택
selected_features = []

for feature in X.columns:
    group_0 = X[y == 0][feature]
    group_1 = X[y == 1][feature]

    t_statistic, p_value = ttest_ind(group_0, group_1, equal_var=False)

    if p_value < 0.05:  # 유의수준 0.05로 설정
        selected_features.append({'Variable': feature, 'p-value': p_value})

# 결과를 데이터프레임으로 만들어 출력
result_df_wt = pd.DataFrame(selected_features)

result_df_wt.reset_index()
result_df_wt
```
