창업자 전공

* 1 : 공학(엔지니어)
    * C, D, E, F, J
* 2 : 자연과학
    * A, B, D, E, M, 
* 3 : 경영/경제학
    * G, K, I, N, R, S, H
* 4 : 인문사회학
    * P, Q,
