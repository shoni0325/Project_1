### 1, 2 -> 0, 1 로 변환

```python
df['상장여부'] = df['상장여부'].replace([2], 0)
df['대표이사_창업자여부'] = df['대표이사_창업자여부'].replace([2], 0)
df['단독창업 여부'] = df['단독창업 여부'].replace([2], 0)
df['창업자 과거창업 경험 여부'] = df['창업자 과거창업 경험 여부'].replace([2], 0)
df['융자(정책)_경험여부'] = df['융자(정책)_경험여부'].replace([2], 0)
df['자체브랜드보유여부'] = df['자체브랜드보유여부'].replace([2], 0)
df['전공과 업종 일치 여부'] = df['전공과 업종 일치 여부'].replace([2], 0)
df['엔젤투자자 혹은 액셀러레이터 투자 경험여부'] = df['엔젤투자자 혹은 액셀러레이터 투자 경험여부'].replace([2], 0)
```

```python
df['인수합병(MnA) 경험여부'] = df['인수합병(MnA) 경험여부'].replace([2], 0)
df['인수합병(MnA) 경험여부'] = df['인수합병(MnA) 경험여부'].replace([3], 2)
```

* 창업 당시 창업자 연령 -> 5
* 제조 및 비제조 -> 4
* 창업 당시 창업자 최종학력 -> 5
* 창업 직전 근무지 -> 10
* 창업방식 -> 3
* 창업 당시 목표시장 -> 3

* 영업및마케팅방식 -> 4
* 연구개발관련전담부서보유여부 -> 4
* 해외수출 및 진출과정 애로사항 -> 8
